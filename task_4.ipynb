```python
from google.colab.patches import cv2_imshow
import cv2
import matplotlib.pyplot as plt
import numpy as np

def gamma(pixel_data, gamma, max_color):
    c = 1
    for i in range(0, len(pixel_data), 3):
        for j in range(3):
            pixel_data[i + j] = int(c * (pixel_data[i + j] / max_color) ** gamma * max_color)
    return pixel_data

img = cv2.imread('DIP_Lab_task_04.jpg')
black_img = gamma(img.flatten(), 8, 255)
black_img = black_img.reshape(img.shape)

plt.figure(figsize=(15, 8))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.title('Normal Image')
plt.subplot(1, 2, 2)
plt.imshow(black_img, cmap='gray')
plt.axis('off')
plt.title('Decreased Brightness Image')
```
