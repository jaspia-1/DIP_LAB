```python
from google.colab.patches import cv2_imshow
import cv2
import matplotlib.pyplot as plt
import numpy as np

def transform(img):
  row, col, channel = img.shape

  img2 = np.zeros((row, col), dtype=np.uint8)

  for i in range(row):
    for j in range(col):
      r, g, b = img[i, j]
      img2[i, j] = r * 0.299 + g * 0.587 + b * 0.114

  return img2

img = cv2.imread('DIP_Lab_task_02.jpg')
img2 = transform(img)

plt.figure(figsize=(5, 5))
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.axis('off')
plt.title('RGB Image')
plt.subplot(1, 2, 2)
plt.imshow(img2, cmap='gray')
plt.axis('off')
plt.title('Gray Scale Image')
```
