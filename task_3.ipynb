```python
# The image contains salt and pepper noise

from google.colab.patches import cv2_imshow
import cv2
import matplotlib.pyplot as plt
import numpy as np

def transform(img, filter):
  xh, xw = filter.shape

  yh = xh//2
  yw = xw//2

  img2 = np.pad(img, ((yh, yh), (yw, yw)), 'constant')
  x, y = img.shape
  res = np.zeros((x, y))

  for i in range(x):
    for j in range(y):
        cur = img2[i : i + xh, j : j + xw]
        res[i, j] = np.median(cur)

  return res

img = cv2.imread('DIP_Lab_task_03.png')
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
karnel = np.zeros((5, 5), dtype=np.uint8)
img2 = transform(gray_img, karnel)

plt.figure(figsize=(15, 8))
plt.subplot(1, 2, 1)
plt.imshow(gray_img, cmap='gray')
plt.axis('off')
plt.title('Gray Scale Image')
plt.subplot(1, 2, 2)
plt.imshow(img2, cmap='gray')
plt.axis('off')
plt.title('Smoothing Image')
```
